The forward pass should be based on the Relu activation function and the activation function should be applied to the final outputs of h1,h2 and y only. Now compute the same forward pass, but for (0,0),(1,0),(1,1)

```python
#Put your code below.
def forward_pass(activation_function):
    # Hidden layer
    #h1_in = 1 * x1 + 1 * x2 - 0.5
    h1_in = 1 * 0 + 1 * 1 - 0.5
    h1_out = activation_function(h1_in)

    #h2_in = -1 * x1 + 1 * x2 - 0.5
    h2_in = -1 * 0 + 1 * 1 - 0.5
    h2_out = activation_function(h2_in)

    # Output layer
    y_in = 1 * h1_out + 1 * h2_out - 0.5
    y_out = activation_function(y_in)

    return h1_in, h1_out, h2_in, h2_out, y_in, y_out
#you need to write the input for h1 and h2, the output for h1 and h2. Input for y and output for y.

forward_pass(ReLu)
```

```python
#Put your code below.
def forward_pass(x1, x2, activation_function):
    # Hidden layer
    h1_in = 1 * x1 + 1 * x2 - 0.5
    h1_out = activation_function(h1_in)

    h2_in = -1 * x1 + 1 * x2 - 0.5
    h2_out = activation_function(h2_in)

    # Output layer
    y_in = 1 * h1_out + 1 * h2_out - 0.5
    y_out = activation_function(y_in)

    return h1_in, h1_out, h2_in, h2_out, y_in, y_out
#you need to write the input for h1 and h2, the output for h1 and h2. Input for y and output for y.

forward_pass(0,0,ReLu)
```
